```python
from sklearn.datasets import load_breast_cancer
import numpy as np
cancer = load_breast_cancer()
X = cancer.data
features = cancer.feature_names
```

(Preprocessing) Wir zentrieren die Daten, so dass $\sum_{i=1}^{569} X_{ic} = 0$ für alle $c \in \{1,..,30\}$,
und bringen sie auf einheitliche Varianz $\sum_{i=1}^{569} (X_{ic})^2 = 1$ für alle $c \in \{1,..,30\}$.

```python
X = X - np.mean(X, axis=0)
X = X / np.sqrt(np.var(X,axis=0))
```

$30$-dimensionale Daten lassen sich nicht visualisieren. Daher suchen wir nach niedrigdimensionalen
Projektionen, die 'so viel Variation wie möglich' enthalten.

Wir versuchen zuerst zwei zufällige Richtungen.

```python
np.random.seed(1234)
vv_1 = np.random.normal(size=30)
vv_2 = np.random.normal(size=30)
print(vv_1)
print(vv_2)
```

Wir orthorormalisieren:

```python
vw_1 = vv_1 / np.sqrt( np.inner( vv_1, vv_1 ) )
```

```python
tmp = vv_2 - np.inner(vw_1, vv_2) * vw_1
vw_2 = tmp / np.sqrt( np.inner(tmp,tmp) )
```

Sie sind nun tatsächlich (fast) orthonormal:

```python
print(np.inner(vw_1,vw_1))
print(np.inner(vw_1,vw_2))
print(np.inner(vw_2,vw_2))
```

```python
zufaellige_koordinate_1 = np.dot( X, vw_1 )
zufaellige_koordinate_2 = np.dot( X, vw_2 )
print(zufaellige_koordinate_1.shape)
print(zufaellige_koordinate_2.shape)
```

```python
import matplotlib.pyplot as plt

def plot(_x,_y):
    cdict = {0:'red',1:'green'}
    labl = {0:'Malignant',1:'Benign'}
    marker = {0:'*',1:'x'}
    alpha = {0:.3, 1:.5}

    fig,ax=plt.subplots(figsize=(7,5))
    fig.patch.set_facecolor('white')
    for l in np.unique(cancer.target):
        ix=np.where(cancer.target==l)
        ax.scatter(_x[ix], _y[ix], c=cdict[l],s=40, label=labl[l],marker=marker[l],alpha=alpha[l])
    plt.legend()


plot(zufaellige_koordinate_1, zufaellige_koordinate_2)
```

Also: machen wir PCA.

```python
M = X.T @ X
lambd, v = np.linalg.eig(M)
```

```python
lambd
```

```python
sortierte_eigenpaare = sorted(list(zip(lambd,v.T)), key=lambda t: t[0], reverse=True)
```

```python
sortierte_eigenpaare[0]
```

```python
sortierte_eigenpaare[1]
```

Wir nehmen die beiden ersten Hauptkomponenten (die Eigenvektoren, die den größten Eigenwerten entsprechen).

```python
vu_1 = sortierte_eigenpaare[0][1]
vu_2 = sortierte_eigenpaare[1][1]
```

Diese sind automatisch orthogonal (warum?). Und numpy hat sie schon normalisiert ausgegeben:

```python
print(np.inner(vu_1,vu_1))
print(np.inner(vu_1,vu_2))
print(np.inner(vu_2,vu_2))
```

```python
pc_1 = np.dot( X, vu_1 )
pc_2 = np.dot( X, vu_2 )
plot(pc_1, pc_2)
```

Zuletzt schauen wir uns an, wie gut die Hauptkomponenten die Varianz der Daten darstellen.


proj_u x  = <x, u> u
proj_u =      u * u^T

```python
def projektion_auf_die_ersten_HK(_q):
    P = np.zeros( (30, 30) )
    for r in range(_q):
        vu = sortierte_eigenpaare[r][1]
        # Mache aus einem Vektor in R^30 eine Matrix in R^{30x1}:
        vu = vu[:,np.newaxis]
        P = P + vu * vu.T
    return P
```

```python
erklaerte_varianz = []
for q in range(1,30+1):
    Pq = projektion_auf_die_ersten_HK(q)
    Xq = np.dot( X, Pq.T )
    var = 0
    for i in range(569):
        var += np.inner(Xq[i], Xq[i])
    erklaerte_varianz.append(var)
```

```python
plt.plot( range(1,30+1), erklaerte_varianz )
```

Wir sehen, dass schon die ersten $\sim 10$ Hauptkomponenten, den Großteil der Gesamtvarianz reproduzieren können.
